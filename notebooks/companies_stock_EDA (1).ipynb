```python
# Structured stock EDA cell: robust loading, summaries, and saved outputs
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

repo_root = os.path.join('..')
data_dir = os.path.join(repo_root, 'Data', 'yfinance_data')
outputs_dir = os.path.join(repo_root, 'outputs')
os.makedirs(outputs_dir, exist_ok=True)

tickers = ['AAPL','AMZN','GOOG','META','MSFT','NVDA']
frames = {}
for t in tickers:
    path_hist = os.path.join(data_dir, f'{t}_historical_data.csv')
    path_simple = os.path.join(data_dir, f'{t}.csv')
    path = None
    if os.path.exists(path_hist):
        path = path_hist
    elif os.path.exists(path_simple):
        path = path_simple
    else:
        print('Missing both patterns for', t, '->', path_hist, 'and', path_simple)
        continue
    df = pd.read_csv(path)
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df = df.set_index('Date')
    frames[t] = df

# Save summary CSVs and quick plots
summary = []
for t, df in frames.items():
    summary.append({'ticker': t, 'shape': df.shape, 'nulls': df.isnull().sum().to_dict()})
    # descriptive stats
    desc = df.describe().transpose()
    desc.to_csv(os.path.join(outputs_dir, f'{t}_describe.csv'))
    df.isnull().sum().to_csv(os.path.join(outputs_dir, f'{t}_nulls.csv'))
    # close price plot (saved)
    if 'Close' in df.columns:
        plt.figure(figsize=(10,4))
        df['Close'].plot(title=f'{t} Close Price')
        plt.tight_layout()
        plt.savefig(os.path.join(outputs_dir, f'{t}_close.png'))
        plt.close()

pd.DataFrame([{'ticker':s['ticker'], 'rows':s['shape'][0], 'cols':s['shape'][1]} for s in summary]).to_csv(os.path.join(outputs_dir, 'stocks_shapes.csv'), index=False)
print('Stock EDA outputs written to', outputs_dir)
```
